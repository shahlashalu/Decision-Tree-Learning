Convert all object type datas into numeric for more efficient analysis and visualization.

```python
df['furnishingstatus'] = df['furnishingstatus'].replace({'furnished':0,'semi-furnished':1,'unfurnished':2})
df.head()
```

```python
df['mainroad'] = df['mainroad'].replace({'no':0,'yes':1})
df['guestroom'] = df['guestroom'].replace({'no':0,'yes':1})
df['basement'] = df['basement'].replace({'no':0,'yes':1})
df['airconditioning'] = df['airconditioning'].replace({'no':0,'yes':1})
df['prefarea'] = df['prefarea'].replace({'no':0,'yes':1})
df['hotwaterheating'] = df['hotwaterheating'].replace({'no':0,'yes':1})
df.head()
```

```python
x = df.drop('prefarea', axis = 1)
y = df['prefarea']
```

```python
x.head()
```

```python
y.head()
```

**DATA EXPLORATION**

```python
from sklearn.feature_selection import SelectKBest
bestfeatures = SelectKBest(k = 11)

fit = bestfeatures.fit(x,y)

fsscores = pd.DataFrame(fit.scores_)

fscolumns = pd.DataFrame(x.columns)

fs = pd.concat([fscolumns,fsscores],axis = 1)
fs.columns = ['plc','Score']

print(fs)
```

```python
fit.scores_
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
y_train.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn import tree
# from sklearn.tree import DecisionTreeClassifier

dt = tree.DecisionTreeClassifier(max_depth = 4, criterion = 'entropy')
dt.fit(x_train,y_train)

y_pred = dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

**MODEL EVALUATION**

```python
testing_score = dt.score(x_test, y_test)
testing_score
```

```python
from sklearn.metrics import confusion_matrix
con_mat = confusion_matrix(y_test, y_pred)
con_mat
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print('MSE : ',mse)
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred, average = 'binary')
print('Precision : %.3f' %precision)
```

```python
from sklearn.metrics import recall_score
recall = recall_score(y_test, y_pred, average = 'binary')
print('Recall : %.3f' %recall)
```

```python
from sklearn.metrics import f1_score
score = f1_score(y_test, y_pred, average = 'binary')
print('F1 Score : %.3f' %score)
```
